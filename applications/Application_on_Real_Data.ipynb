```python
import pandas as pd
df = pd.read_excel('mnist.xlsx', index_col=None, header=0)  
df
```

```python
from matplotlib import pyplot as plt
df1 = df[50:][['sghmc', 'sgld','nag']]
```

```python
plt.figure(figsize = (12,8))
plt.plot(df1['sghmc'], 'r-.',  label = 'SGHMC')
plt.plot(df1['nag'], 'b--', label = 'SGD with momentum')
plt.plot(df1['sgld'], 'c--', label = 'SGLD')
plt.legend(loc = 1)
plt.ylabel('Test error')
plt.xlabel('Iteration')
plt.savefig("Real_data_1.png")
```
